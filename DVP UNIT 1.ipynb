```python
import cv2
import os
from tqdm import tqdm
file = "C:\\Users\\kishore\\Downloads\\Testvideo.mp4"
video = cv2.VideoCapture(file)
while video.isOpened():
    #READING EACH FRAME
    ret, frame = video.read()
    if not ret:
        print("End of video")
        break
 #SHOWING EACH FRAME
    cv2.imshow("Video_name", frame)
 #IF KEY q IS PRESSED, EXIT THE VIDEO
    if cv2.waitKey(1) == ord("q"):
        print("Program break...")
        break
video.release()
cv2.destroyAllWindows()
```

```python
#exp-3
import cv2
import os
from tqdm import tqdm
import cv2
from matplotlib import pyplot as plt
import os
from tqdm import tqdm
#READING IMAGES FROM WHICH FRAMES ARE TO BE CREATED
im1 = cv2.imread("C:\\Users\\kishore\\Pictures\\strawberry.png")
im1=cv2.cvtColor(im1, cv2.COLOR_BGR2RGB)
im2 = cv2.imread("C:\\Users\\kishore\\Pictures\\freefire1.jpg")
im2=cv2.cvtColor(im2, cv2.COLOR_BGR2RGB)
im3 = cv2.imread("C:\\Users\\kishore\\Pictures\\Saved Pictures\\OSJAVAPIC.jpg")
im3=cv2.cvtColor(im3, cv2.COLOR_BGR2RGB)
im4 = cv2.imread("C:\\Users\\kishore\\Pictures\\Screenshots\\Screenshot 2023-05-28 175148.png")
im4=cv2.cvtColor(im4, cv2.COLOR_BGR2RGB)
im2=cv2.resize(im2, (im1.shape[1], im1.shape[0]))
im3=cv2.resize(im3, (im1.shape[1], im1.shape[0]))
im4=cv2.resize(im4, (im1.shape[1], im1.shape[0]))
fig,ax = plt.subplots(nrows = 2, ncols=2, figsize=[20,20])
ax[0][0].imshow(im1)
ax[0][1].imshow(im2)
ax[1][0].imshow(im3)
ax[1][1].imshow(im4)
fig.tight_layout()
plt.show()
```

```python
import cv2
from tqdm import tqdm

path = "C:\\Users\\kishore\\Pictures\\Saved Pictures"
for i in tqdm(range(0,120)):
    alpha = round(i/120, 3)
    tmp = cv2.addWeighted(im1, 1-alpha, im2, alpha,0)
    cv2.imwrite(path + str(i) + '.jpg', tmp)
for i in tqdm(range(120,240)):
    alpha = round((i-120)/120, 3)
    tmp = cv2.addWeighted(im2, 1-alpha, im3, alpha,0)
    cv2.imwrite(path + str(i) + '.jpg', tmp)
for i in tqdm(range(240,360)):
    alpha = round((i-240)/120, 3)
    tmp = cv2.addWeighted(im1, 1-alpha, im2, alpha,0)
    cv2.imwrite(path + str(i) + '.jpg', tmp)
```

```python
import cv2
import os
from tqdm import tqdm

imgs = []
path = "C:\\Users\\kishore\\Pictures\\Saved Pictures"

for fl in tqdm(os.listdir(path)):
    img_path = os.path.join(path, fl)
    img = cv2.imread(img_path)
    if img is not None:
        imgs.append(img)
print(str(len(imgs)) + ' images loaded')
```

```python
import cv2
out = cv2.VideoWriter("C:\\Users\\kishore\\Desktop\\DVP\\DVP VIDEO FILE.mp4",cv2.VideoWriter_fourcc(*'DIVX'), 25, (imgs[0].shape[1], imgs[0].shape[0]))
for im in tqdm(imgs):
    out.write(im)
out.release()
```
